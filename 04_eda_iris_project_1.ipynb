# Conclution:

1. According to the pair plot, it is concluded that Petal length and petal width is the most importat features to identify 
flower types.

Petal length 0-2 & petal width 0-0.7 => Setosa 
Petal length 3-5.2 & petal width 1-1.7 => Versicolor 
Petal length less than 5.2 & petal width is greater than 1.7 => Verginica

2. By the plotsSetosa can be easily identified (linearly sepratable), verginica and versicolor have some overlap 
(almost linearly separatble)

3. We can find lines and by using if-else conditions we can build a simple model to classify flower type.
